##  analise dos dados

```python
import pandas as pd
```

```python
df = pd.read_csv("../datasets/credit_01/test")
df.head()
```

Os dados contem dados bastante heterogeneos. Observamos strings, inteiros, caracteres e etc... sem serem propriamente indentificados. Primeiramente separemos os dados compostos por numeros e outros tipos de dados.

```python
df_num = pd.DataFrame()
df_obj = pd.DataFrame()
for colum in df.columns:
    if df[colum].dtype!="object":
        df_num[colum]= df[colum]
    else:
        df_obj[colum]= df[colum]
for colum in df_num.columns:
    print(df[colum].dtype)

print("******************************************************************************")
for colum in df_obj.columns:
    print(df[colum].dtype)

```

```python
print("num of number "+ str(df_num.columns.size))
print("num of objects "+ str(df_obj.columns.size))
```

vemos que o numero de dados composto por numeros é 37, e o de objetos são 114

```python
df_num.to_csv("nums_param.csv")
df_obj.to_csv("obj_param.csv")
```
